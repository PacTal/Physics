# Trayectoria de partículas en un camo elé|ctrico y magnético

```
# Paco H. Talero L.
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
import numpy as np
from pylab import*
```

```
#---Parámetros---
DeltaT=0.001 #---Delta del tiempo (ns)
Tmax=28.0  # tiempo máximo (ns)
#----Dos sistemas------
m,m1=4.0,2.0       # Masa en unidades adecuadas u.m.a (1,67X10^-27Kg)
q,q1=2.0,2.0       # Carga Columbios
Ey,Ey1=0.5,-0.5  # Campo electrico en y N/C
B,B1=1.0,1.0    # Campo magnético + si sale de lo contrario - entra Gauss (10^-4T)
#----Variables en su condicion inicial---
t=0.  #---Tiempo
x,x1=0.,0.  #---posición en x (cm)
y,y1=0.,0.  #---posición en y (cm)
vx,vx1=1.5,-1.5 #---velocidad en vx (10^5 m/s)
vy,vy1=0.0,0.0 #---velocidad en vy

#--Listas---
X,Y,X1,Y1=[],[],[],[]    #---Listas
while(t<=Tmax):
	X.append(x)
	X1.append(x1)  #---partícula 1
	Y.append(y)
	Y1.append(y1)  #---partícula 1
	ax=(q/m)*(vy*B)
	ax1=(q1/m1)*(vy1*B1) #---partícula 1
	ay=(q/m)*(Ey-vx*B)
	ay1=(q1/m1)*(Ey1-vx1*B1) #---partícula 1
	vx=vx+ax*DeltaT
	vx1=vx1+ax1*DeltaT  #---partícula 1
	vy=vy+ay*DeltaT
	vy1=vy1+ay1*DeltaT  #---partícula 1
	x=x+vx*DeltaT
	x1=x1+vx1*DeltaT   #---partícula 1
	y=y+vy*DeltaT
	y1=y1+vy1*DeltaT  #---partícula 1
	t=t+DeltaT
#---Graficos--------------
plot(X,Y,color="red",label='$S$')
plot(X1,Y1,color="blue",label='$S_1$') #---gráfico de la particula 1
#plt.grid(True)
plt.legend(loc='lower left') #---lower center, lower right, upper left
ylabel('$y(cm)$',fontsize=16)
xlabel('$x(cm)$',fontsize=16)
plt.ylim(-6, 6)  # rango
#plt.xlim(0.,80.)  # rango
#plt.axes().set_aspect('equal')
#plt.axis('equal')  # Escalas iguales en ambos ejes
title('Trayectoria')
show()
```
