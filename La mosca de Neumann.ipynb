<h1>El problema de la mosca vagabunda</h1>

Se extiende el problema tradicional que pregunta por la distancia que recorre una mosca con rapidez constante cuando se mueve entre dos trenes en <b>MUR</b> con dirección de colisión. Se explora este problema mediante un algoritmo simple y de gran riqueza didáctica para estudiantes de física elemental,  pues muestra un contexto sobre el cual efectuar un primer acercamiento al análisis gráfico en cinemática.

<a href="https://www.scielo.br/pdf/rbef/v35n2/13.pdf">Leer artículo</a>

```python
import matplotlib.pyplot as plt
```

<h2>Parámetros del sistema</h2>
<ul>
<li>$T_{max}$, tiempo máximo para el experimento virtual.</li>
<li> $\Delta t$, incremento temporal.</li>
<li> $v_m$, rapidez de la mosca.</li>   
</ul>

```python
Tmax,DeltaT,vm=50., 0.01,50.
```

<h2>Condiciones iniciales</h2>
<ul>
<li>$v_a$, velocidad inicial del tren $A$.</li>
<li>$v_b$, velocidad inicial del tren $B$.</li>
<li>$x_{a\:o}$, posición inicial del tren $A$.</li>
<li>$x_{b\:o}$, posición  inicial del tren $B$.</li>
<li>$x_{m\:o}$, posición  inicial de la mosca $B$.</li>
</ul>

```python
va,vb,xao,xbo,xmo=10.,-10.,0.,200.,100
```

<h1>Algoritmo</h1>

Mientras los trenes se acercan entre sí con <b>MUR</b> la mosca se dirige a alguno de ellos, cuando lo toca cambia la dirección de su velocidad manteniendo su rapidez y  se dirige al otro con <b>MUR</b>, lo toca y cambia de nuevo el signo de su velocidad y se nuevo se dirige al primer tren. La mosca continúa moviéndose entre cada tren hasta que estos chocan y la “aplastan”. En el proceso los trenes y la mosca actualizan sus posiciones con $x\leftarrow x+v\Delta t$.


```python
T,XA,XB,XM=[],[],[],[]
t,xa,xb,xm=0.,0.,100.,0.
while t<= Tmax:
  T.append(t)
  XA.append(xa)
  XB.append(xb)
  XM.append(xm)
  xa=xa+va*DeltaT
  xb=xb+vb*DeltaT
  xm=xm+vm*DeltaT
  if xm>=xb:
    vm=-vm
    xm=xm+vm*DeltaT
  if xm<=xa:
    vm=-vm
    xm=xm+vm*DeltaT
  if xa>=xb:break
  t=t+DeltaT
```

<h2>Gráficas</h2> 

Se muestran gráficos de posición contra tiempo de la mosca y los trenes. 


```python
plt.plot(T,XA,color="red",label='Tren A')
plt.plot(T,XB,color="blue",label='Tren B')
plt.plot(T,XM,color="green",label='Mosca')
plt.grid(False)
#plt.ylim(0,800)  # rango
#plt.xlim(0,7)  # rango
plt.legend(loc='upper right')  #---lower center, lower right, upper left
plt.ylabel('$x(m)$',fontsize=16)
plt.xlabel('$t(s)$',fontsize=16)
plt.title('La Mosca de Neumann')
plt.show()
```
