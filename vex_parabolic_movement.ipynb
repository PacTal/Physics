```
#--PacTal
import numpy as np
from pylab import*
#---Parámetros---
DeltaT=0.001 #---Delta del tiempo
Th   = np.pi/4.  #--Radianes
Tmax=4.7 # tiempo máximo
m= 0.025     #--Masa en Kg
R=0.019    #--Metros
d=1.2    #--Densidad del medio
C=0.5      #--Coeficiente de arastre
g=9.8      #--Aceleración gravedad
A=np.pi*R**2. #--Área transversal
Vol=(4./3.)*np.pi*R**3. #--Volumen
Fe=d*g*Vol  #--Fuerza de flotación
```

```
#-Condiciones iniciales
xo=0.  #---posición en x metros
yo=20.  #---posición en y metros
v=20.  #---rapidez inicial m/s
```

```
#--Aceleración
def a(vx,vy):
  rap=np.sqrt(vx**2.+vy**2)**0.5
  ax=-0.5*d*A*C*rap*vx/m
  ay=-0.5*d*A*C*rap*vy/m+Fe/m-g
  return ax,ay
```

```
#---programa principal
Tiempo,X,Y,Xl,Yl=[],[],[],[],[] #--posiciones fricción y libres
t,x,y=0.,xo,yo
vx,vy,=v*np.cos(Th),v*sin(Th)
xl,yl=xo,yo #_l is free
while t<=Tmax:
  X.append(x),Y.append(y),Xl.append(xl),Yl.append(yl),Tiempo.append(t)
  xl=xo+v*np.cos(Th)*t
  yl=yo+v*sin(Th)*t-0.5*g*t**2.
  x=x+vx*DeltaT
  y=y+vy*DeltaT
  vx=vx+a(vx,vy)[0]*DeltaT
  vy=vy+a(vx,vy)[1]*DeltaT
  #if y<=0.: break
  t=t+DeltaT
```

```
#--Graficos de Trayectoria
#---Graficos--------------
plot(X,Y,color="red",label='Medio')
plot(Xl,Yl,color="blue",label='Libre') #---grfico de la particula libre
plt.grid(True)
plt.legend(loc='upper right') #---lower center, lower right, upper left
ylabel('$y(m)$',fontsize=16)
xlabel('$x(m)$',fontsize=16)
#plt.ylim(0, 25)  # rango
#plt.xlim(0.,80.)  # rango
title('Trayectoria')
show()
```
